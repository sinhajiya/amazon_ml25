```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
df = r"D:\amazon_dataset\student_resource\dataset\train.csv"
df = pd.read_csv(df)
```

```python
df.head()
```

```python
df.info()
```

Visualizing prices distributions

```python
prices = df['price'].astype(float)

plt.figure(figsize=(10, 6))
sns.kdeplot(prices, fill=True, alpha=0.5, linewidth=2, color="skyblue")
plt.title("Distribution of Raw Prices", fontsize=14)
plt.xlabel("Price", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("raw_price_distribution.png", dpi=300)
plt.show()



```

The data is highly skewed. there are relatively few samples for high prices.

```python
log_prices = np.log1p(prices)  

plt.figure(figsize=(10, 6))
sns.kdeplot(log_prices, fill=True, alpha=0.5, linewidth=2, color="salmon")
plt.title("Distribution of Log(1 + Price)", fontsize=14)
plt.xlabel("Log(1 + Price)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("log_price_distribution.png", dpi=300)
plt.show()
```
